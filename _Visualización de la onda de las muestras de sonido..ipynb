# Gr√°ficas de Piano 

```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import wave, sys
raw = wave.open("piano-G3.wav")
signal = raw.readframes(-1)
signal = np.frombuffer(signal, dtype ="int16")
f_rate = raw.getframerate()
time = np.linspace(
	0, # start
	len(signal) / f_rate,
	num = len(signal)
)
plt.figure(1)
plt.title("Sound Wave")
plt.xlabel("Time")
plt.plot(time, signal)
plt.show()
```
